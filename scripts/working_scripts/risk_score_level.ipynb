```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
out=pd.read_csv('LA_daily_out.csv')
print(out.head(5))
```

```python
# USC's team defines: risk score=(number of infectious people × effective reproductive number × 10,000)/population size
# Umich's team defines: risk score=(number of infectious people × 10,000)/population size
# USC's team: 
# 1-low level(risk score<0.1);
# 2-low/medium level(risk score 0.1-1);
# 3-medium/high level(risk score 1-2); 
# 4-high level(risk score >2)
```

```python
# July
Jul26=out.loc[out['Timestamp']=='2020-07-26 ',['Region','Risk_score_level','Risk_score']]
print(Jul26.head())

# histogram of risk score
plt.hist(Jul26['Risk_score'],range=(0,10),facecolor="blue", edgecolor="black", alpha=0.7,rwidth=0.8)
plt.title("07/26 risk score")
plt.show()

# bar chart of risk score level(according to USC's)
# 1-low level(<0.1);2-low/medium level(0.1-1);3-medium/high level(1-2); 4-high level(>2)
plt.hist(Jul26['Risk_score_level'],facecolor="blue", edgecolor="black", alpha=0.7,rwidth=0.8)
plt.title("07/26 risk level")
plt.show()
```
